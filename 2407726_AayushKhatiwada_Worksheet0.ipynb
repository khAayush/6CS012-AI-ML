of the numbers).

```python
def find_sum(numbers):
    """
    Calculates the sum of a list of numbers.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: The sum of the numbers.
    """
    return sum(numbers)


def find_average(numbers):
    """
    Calculates the average of a list of numbers.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: The average of the numbers.
    """
    return sum(numbers) / len(numbers)


def find_maximum(numbers):
    """
    Finds the maximum value in a list of numbers.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: The maximum number in the list.
    """
    return max(numbers)


def find_minimum(numbers):
    """
    Finds the minimum value in a list of numbers.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: The minimum number in the list.
    """
    return min(numbers)


print("Mathematical Operations Program")
print("------------------------------")
print("1. Find Sum")
print("2. Find Average")
print("3. Find Maximum")
print("4. Find Minimum")

try:
    choice = int(input("Enter your choice (1-4): "))
    user_input = input("Enter numbers separated by spaces: ")
    numbers = list(map(float, user_input.split()))

    if not numbers:
        raise ValueError("List cannot be empty.")

    if choice == 1:
        print("Sum:", find_sum(numbers))
    elif choice == 2:
        print("Average:", find_average(numbers))
    elif choice == 3:
        print("Maximum:", find_maximum(numbers))
    elif choice == 4:
        print("Minimum:", find_minimum(numbers))
    else:
        print("Error: Invalid operation selected.")

except ValueError:
    print("Error: Please enter valid numeric values and make sure the list is not empty.")
```

#**3. Exercise on Nested List:**

## 3.1. Flatten a Nested List:

Write a Python function that takes a nested list and flattens it into a single list, where all the elements are in a single dimension.





```python
def flatten(lst):

    flat = []
    for sublist in lst:
        for item in sublist:
            flat.append(item)
    return flat

nested_list = [[1, 2], [3, 4], [5]]
print(flatten(nested_list))
```

## 3.2. Accessing Nested List Elements:

Write a Python function that extracts a specific element from a nested list given its indices.



```python
def access_nested_element(lst, indices):

    element = lst
    for index in indices:
        element = element[index]
    return element

nested_list = [[1,2,3],[4,5,6],[7,8,9]]
print(access_nested_element(nested_list, [1,2]))
```

## 3.3. Sum of All Elements in a Nested List:

Write a Python function that calculates the sum of all the numbers in a nested list (regardless of depth).



```python
def sum_nested(lst):

    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total

nested_list = [[1, 2], [3, [4, 5]], 6]
print(sum_nested(nested_list))
```

## 3.4. Remove Specific Element from a Nested List:

Write a Python function that removes all occurrences of a specific element from a nested list.


```python
def remove_element(lst, elem):

    result = []
    for item in lst:
        if isinstance(item, list):
            result.append(remove_element(item, elem))
        elif item != elem:
            result.append(item)
    return result

nested_list = [[1, 2], [3, 2], [4, 5]]
print(remove_element(nested_list, 2))
```

## 3.5. Find the Maximum Element in a Nested List:

Write a Python function that finds the maximum element in a nested list (regardless of depth).



```python
def find_max(lst):

    maximum = float('-inf')
    for item in lst:
        if isinstance(item, list):
            maximum = max(maximum, find_max(item))
        else:
            maximum = max(maximum, item)
    return maximum

nested_list = [[1, 2], [3, [4, 5]], 6]
print(find_max(nested_list))
```

## 3.6. Count Occurrences of an Element in a Nested List:

Write a Python function that counts how many times a specific element appears in a nested list.



```python
def count_occurrences(lst, elem):
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        elif item == elem:
            count += 1
    return count


nested_list = [[1, 2], [2, 3], [2, 4]]
print(count_occurrences(nested_list, 2))
```

## 3.7. Flatten a List of Lists of Lists:

Write a Python function that flattens a list of lists of lists into a single list, regardless of the depth.



```python
def deep_flatten(lst):

    flat = []
    for item in lst:
        if isinstance(item, list):
            flat.extend(deep_flatten(item))
        else:
            flat.append(item)
    return flat


nested_list = [[[1,2],[3,4]],[[5,6],[7,8]]]
print(deep_flatten(nested_list))
```

## 3.8. Nested List Average:

Write a Python function that calculates the average of all elements in a nested list.

```python
def average_nested(lst):

    flat_list = deep_flatten(lst)
    return sum(flat_list) / len(flat_list)
nested_list = [[1,2],[3,4],[5,6]]
print(average_nested(nested_list))
```

### Problem - 4: Matrix Operations:
• For the following arrays:

  A = np.array([[3,4],[7,8]]) and B = np.array([[5,3],[2,1]]);

Prove following with Numpy:
1. Prove A.A−1 = I.
2. Prove AB ̸= BA.
3. Prove (AB)

T = BTAT .

• Solve the following system of Linear equation using Inverse Methods.

2x − 3y + z = −1

x − y + 2z = −3

3x + y − z = 9

{Hint: First use Numpy array to represent the equation in Matrix form. Then Solve for: AX = B}

• Now: solve the above equation using np.linalg.inv function.{Explore more about ”linalg” func-
tion of Numpy}

```python
import numpy as np

A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

print("A:\n", A)
print("B:\n", B)
print("\n")

# 1️ Prove A * A^-1 = I
A_inv = np.linalg.inv(A)
identity = np.dot(A, A_inv)
print("1. A * A^-1 = I ?")
print(identity)
print("Check close to identity:", np.allclose(identity, np.eye(2)))  # True if close
print("\n")

# 2️ Prove AB != BA
AB = np.dot(A, B)
BA = np.dot(B, A)
print("2. AB =\n", AB)
print("   BA =\n", BA)
print("Check AB == BA:", np.allclose(AB, BA))  # False, so AB != BA
print("\n")

# 3️ Prove (AB)^T = B^T * A^T
AB_T = AB.T
B_TA_T = np.dot(B.T, A.T)
print("3. (AB)^T =\n", AB_T)
print("   B^T * A^T =\n", B_TA_T)
print("Check equality:", np.allclose(AB_T, B_TA_T))
print("\n")


# Represent in matrix form AX = B
A_eq = np.array([[2, -3, 1],
                 [1, -1, 2],
                 [3, 1, -1]])
B_eq = np.array([-1, -3, 9])

# Using np.linalg.inv
A_eq_inv = np.linalg.inv(A_eq)
X = np.dot(A_eq_inv, B_eq)
print("Solution using inverse method:")
print("x, y, z =", X)

# 3. Alternative: Using np.linalg.solve

X_solve = np.linalg.solve(A_eq, B_eq)
print("Solution using np.linalg.solve (recommended):")
print("x, y, z =", X_solve)
```
